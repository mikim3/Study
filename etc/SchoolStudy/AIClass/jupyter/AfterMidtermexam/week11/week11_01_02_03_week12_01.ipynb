```python
# 결측치가 있는 행 제거
data_del_row = data.dropna(axis = 0)

data_del_row
```

```python
# 결측치 대체하는 방법
# fillna()
# fill  <---  빈칸을 채우다
# NA
#  1) 특정 값으로 대체
#  2) 각 열 별 평균값으로 대체
#  3) 다른 열의 값으로 대체
#  4) 그룹 평균값으로 대체
```

```python
# 1) 특정 값으로 대체
data
```

```python
data_0 = data.fillna(0)
data_0
```

```python
# Missing Value
data_missing = data.fillna('missing')
data_missing
```

```python
data_ffill = data.fillna(method='ffill')   # front fill
data_ffill
```

```python
data_bfill = data.fillna(method='bfill')   # back fill
data_bfill
```

```python
#  2) 각 열 별 평균값으로 대체
#  3) 다른 열의 값으로 대체
#  4) 그룹 평균값으로 대체
```

```python
# 2) 각 열 별 평균값으로 대체
data.mean()
```

```python
data_mean = data.fillna(data.mean())
data_mean
```

```python
data_min = data.fillna(data.min())  # 최솟값 대체
data_min
```

```python
data.max()
```

```python
data_max = data.fillna(data.max())  # 최댓값 대체
data_max
```

```python
#  3) 다른 열의 값으로 대체
#  4) 그룹 평균값으로 대체
```

```python
# 3) 다른 열의 값으로 대체
# sugar 열의 결측치를 alcohol열의 값으로 대체

import numpy as np

data2 = data.copy()

# not null
# 결측값 Yes ---> False
# 결측값 No  ---> True
data2['new_sugar'] = np.where(pd.notnull(data2['sugar']) == True, data2['sugar'], data2['alcohol'])
data2
```

```python
#  4) 그룹 평균값으로 대체
data
```
